Решим задачу классификации баскетбольных команд по полу при помощи Random Forest.
Обучаем Random forest классификатор с параметром n_estimators = 10 (n_estimators = количество построенных деревьев).

```python
from sklearn import ensemble
rf = ensemble.RandomForestRegressor(max_depth=10, random_state=0)
rf.fit(X_train, y_train)
#Проверим качество классификатора по величине ошибки на тестовой и обучающей выборках
y_train_pred = rf.predict(X_train)
y_test_pred = rf.predict(X_test)
train_err.append(np.square(y_train-y_train_predict).mean())
test_err.append(np.square(y_test-y_test_predict).mean())
print("Ошибка на обучающей выборке: ", err_train)
print("Ошибка на тестовой выборке: ", err_test)
```


Наблюдается ситуация переобучения, т.к. ошибка на тестовой выборке оказалась выше, чем на обучающей

Построим график зависимости ошибки на тестовой и обучающей выборках и oob-ошибки от количества деревьев. Установим значение парметра oob_score = True, данный параметр определяет использовать ли out-of-bag объекты для оценки

Рассмотрим график на диапазоне от 1 до 10.

```python
tree_arr = np.arange(1, 10, 1)
test_err = []
train_err = []
oob_err = []

for tree in tree_arr:
    rf = ensemble.RandomForestRegressor(max_depth=10, random_state=0)
    rf.fit(X_train, y_train)
    y_train_pred = rf.predict(X_train)
    y_test_pred = rf.predict(X_test)
    train_err.append(np.square(y_train-y_train_predict).mean())
    test_err.append(np.square(y_test-y_test_predict).mean())
    oob_err.append(1 - rf.oob_score)
```

```python
plt.plot(tree_arr, train_err, 'b-o', label = 'train')
plt.plot(tree_arr, test_err, 'r-o', label = 'test')
plt.plot(tree_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.min(tree_arr), np.max(tree_arr)])
plt.title('Error vs. Count of tree')
plt.xlabel('Count of tree')
plt.ylabel('Error')
plt.legend()
pass
```

```python
На всем диапазоне значений наблюдается ситуация переобучения, также видим, что oob ошибка 
превосходит ошибку на тестовой и обучающей выборках, причем с увеличением количества деревьев, 
значение oob-ошибки постепенно уменьшается. При 3 деревьях ~ 0.33, при 6 ~ 0.3, при 9 ~ 0.17
Рассмотрим график на диапазоне от 10 до 100.tree_arr = np.arange(10, 100, 5)
```

```python
test_err = []
train_err = []
oob_err = []

for tree in tree_arr:
    rf = ensemble.RandomForestRegressor(max_depth=10, random_state=0)    
    rf.fit(X_train, y_train)
    y_train_pred = rf.predict(X_train)
    y_test_pred = rf.predict(X_test)
    train_err.append(np.square(y_train-y_train_predict).mean())
    test_err.append(np.square(y_test-y_test_predict).mean())
    oob_err.append(1 - rf.oob_score)
```

Также можно сказать, что при малом количестве деревьев (10-20) oob ошибка в целом выше, чем при более большом (30-100). На всем рассматриваемом диапазоне количества деревьев, oob-ошибка превышает (кроме случая n_estimators = 30) значения ошибок на тестовой и обучающей выборках. Минимальное значение oob ошибки достигается при количестве деревьев равном 30

## Extremely Randomized Trees ##
Решим задачу классификации баскетбольных команд по полу при помощи Extremely Randomized Trees.

Обучаем Extremely Randomized Trees классификатор с параметром n_estimators = 10 (n_estimators = количество построенных деревьев).

```python
ert = ensemble.ExtraTreesRegressor(n_estimators = 10)
ert.fit(X_train, y_train)
y_train_pred = ert.predict(X_train)
y_test_pred = ert.predict(X_test)
train_err.append(np.square(y_train-y_train_predict).mean())
test_err.append(np.square(y_test-y_test_predict).mean())
print("Ошибка на обучающей выборке: ", err_train)
print("Ошибка на тестовой выборке: ", err_test)
```

Ошибка на тестовой выборке превосходит ошибку на обучающей. Наблюдается ситуация переобучения.

Построим график зависимости ошибки на тестовой и обучающей выборках и oob-ошибки от количества деревьев. Установим значения параметров oob_score = True (Определяет использовать ли out-of-bag объекты для оценки), bootstrap = True (Определяет применять ли бустрэп для построения дерева)

Рассмотрим график на диапазоне от 1 до 10.

```python
tree_arr = np.arange(1, 10, 1)
test_err = []
train_err = []
oob_err = []
for tree in tree_arr:
    ert = ensemble.ExtraTreesRegressor(n_estimators = tree, oob_score=True, bootstrap=True)
    ert.fit(X_train, y_train)
    train_err.append(np.square(y_train-y_train_predict).mean())
    test_err.append(np.square(y_test-y_test_predict).mean())
    oob_err.append(1 - ert.oob_score)
```

```python
plt.plot(tree_arr, train_err, 'b-o', label = 'train')
plt.plot(tree_arr, test_err, 'r-o', label = 'test')
plt.plot(tree_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.min(tree_arr), np.max(tree_arr)])
plt.title('Error vs. Count of tree')
plt.xlabel('Count of tree')
plt.ylabel('Error')
plt.legend()
pass
```


На всем диапазоне значений, кроме количества деревьев равном 8, наблюдается ситуация переобучения, также видим, что oob ошибка превосходит ошибку на тестовой и обучающей выборках, причем с увеличением количества деревьев, значение oob-ошибки постепенно уменьшается. При 2 деревьях ~ 0.34, при 6 ~ 0.24, при 9 ~ 0.13

Рассмотрим график на диапазоне от 10 до 100.

```python
tree_arr = np.arange(10, 100, 5)
test_err = []
train_err = []
oob_err = []
for tree in tree_arr:
    ert = ensemble.ExtraTreesRegressor(n_estimators = tree, oob_score=True, bootstrap=True)
    ert.fit(X_train, y_train)
    train_err.append(np.square(y_train-y_train_predict).mean())
    test_err.append(np.square(y_test-y_test_predict).mean())
    oob_err.append(1 - ert.oob_score)
```

```python
plt.plot(tree_arr, train_err, 'b-o', label = 'train')
plt.plot(tree_arr, test_err, 'r-o', label = 'test')
plt.plot(tree_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.min(tree_arr), np.max(tree_arr)])
plt.title('Error vs. Count of tree')
plt.xlabel('Count of tree')
plt.ylabel('Error')
plt.legend()
pass
```


При количестве деревьев ∈ [10, 15) наблюдается ситуация переобучения, для количества деревьев не принадлежащих данному интервалу, переобучение отсутствует. На всем рассматриваемом диапазоне количества деревьев oob-ошибка превышает или равна значению ошибок на тестовой и обучающей выборках
